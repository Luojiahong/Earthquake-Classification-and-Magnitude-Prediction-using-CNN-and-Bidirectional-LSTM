# ETS Decomposition(Analysis Model)

```python
import pandas as pd
ets_data = pd.read_csv('metadata.csv')
```

```python
ets_data = ets_data[ets_data['trace_category'] == 'earthquake_local']
```

```python
temp_df=ets_data[['receiver_latitude','receiver_longitude','trace_start_time','source_magnitude']]
temp_df=temp_df[(temp_df['receiver_latitude'] == self_latitude) & (temp_df['receiver_longitude'] == self_longitude)]
temp_df=temp_df[['trace_start_time','source_magnitude']]
temp_df=temp_df.set_index('trace_start_time')
temp_df=temp_df.dropna()
temp_df
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
def etsDecomposition(temp_df):
    temp_df.reset_index(inplace=True)
    temp_df['trace_start_time'] = pd.to_datetime(temp_df['trace_start_time'])
    temp_df = temp_df.set_index('trace_start_time')
    decomposition = seasonal_decompose(temp_df.source_magnitude,freq=30)
    #ets_result=decomposition.plot()
    fig, (ax1,ax2,ax3,ax4) = plt.subplots(4,1, figsize=(15,15))
    ax1.set_title('ETS Decompostion')
    ax1.set(ylabel='Observed')
    ax2.set(ylabel='Trend')
    ax3.set(ylabel='Seasonal')
    ax4.set(ylabel='Residual')
    decomposition.observed.plot(ax=ax1)
    decomposition.trend.plot(ax=ax2)
    decomposition.seasonal.plot(ax=ax3)
    decomposition.resid.plot(ax=ax4)
    plt.savefig("figure.png")
```

```python
etsDecomposition(temp_df)
```
