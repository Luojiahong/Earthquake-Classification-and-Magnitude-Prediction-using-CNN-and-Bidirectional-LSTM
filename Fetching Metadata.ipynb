### Earthquake Magnitudes

<p>Nano earthquake when 1 < M</p>
<p>Micor earthquake when 1 <= M < 3 </p>
<p>Small earthquake when 3 <= M < 5 </p>
<p>moderate when 5 <= M < 7 </p>
<p>major when 7 <= M < 9 </p>

<p>major: 3</p>

```python
sliced_eq_meta = pd.DataFrame(columns=eq_meta.columns)
sliced_eq_meta = sliced_eq_meta.append(eq_meta[(eq_meta['source_magnitude'] > 6) & (eq_meta['source_magnitude'] < 7)].iloc[0:4:2], ignore_index=True)
```

```python
#This code creates sliced_eq_meta dataframe and appends to it 60 nano, 90 micro,  32 Small earthquakes, 15 moderate and 3 major

lower_limit = 0
#sliced_eq_meta = pd.DataFrame(columns=eq_meta.columns)

while(lower_limit <= 7):
    
    #for nano earthquakes
    if lower_limit == 0:
        sliced_eq_meta = sliced_eq_meta.append(eq_meta[eq_meta['source_magnitude']  < 1].iloc[0:120:2], ignore_index=True)
        lower_limit = lower_limit + 1
        
    #for micro earthquakes
    elif lower_limit == 1:
        sliced_eq_meta = sliced_eq_meta.append(eq_meta[(eq_meta['source_magnitude']  >= 1) & (eq_meta['source_magnitude']  < 3)].iloc[0:180:2], ignore_index=True)
        lower_limit = lower_limit + 2
    
    #for small magnitude earthquakes
    elif lower_limit == 3:
        sliced_eq_meta = sliced_eq_meta.append(eq_meta[(eq_meta['source_magnitude']  >= 3) & (eq_meta['source_magnitude']  < 5)].iloc[0:64:2], ignore_index=True)
        lower_limit = lower_limit + 2
        
    #for moderate magnitude earthquakes
    elif lower_limit== 5:
        sliced_eq_meta = sliced_eq_meta.append(eq_meta[(eq_meta['source_magnitude']  >= 5) & (eq_meta['source_magnitude']  < 7)].iloc[0:30:2], ignore_index=True)
        lower_limit = lower_limit + 2
        
    #for major earthquakes
    else:
        sliced_eq_meta = sliced_eq_meta.append(eq_meta[(eq_meta['source_magnitude']  >= 7) & (eq_meta['source_magnitude']  < 9)].iloc[:], ignore_index=True)
        lower_limit = lower_limit + 2
```

```python
#Saves sliced_eq_meta to a csv file
sliced_eq_meta.to_csv('metadata//eq_meta.csv', index=False)
```
